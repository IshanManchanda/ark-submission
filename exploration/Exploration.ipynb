```python
# %%timeit
with open('data/Data1.csv') as csv_file:
	data = np.array(list(csv.reader(csv_file)), dtype=np.float64)
```

```python
# print(my_data.shape)
print(data.shape)
```

```python
ax = plt.axes(projection='3d')
x = 1000
i = 0
ax.scatter3D(data[:x, i], data[:x, i + 1], data[:x, i + 2], c=range(x), cmap='Greens')
```
